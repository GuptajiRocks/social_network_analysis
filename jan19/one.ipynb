```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
edges = [(1, 2), (1,3), (1,4), (2, 3), (2, 4), (2, 5), (3, 4), (3, 5), (3, 6), (4, 5), (4, 6), (4, 7), (5, 6), (5, 7), (5, 8), (6, 7), (6, 8), (7, 8), (5, 1)]

G = nx.Graph()
G.add_edges_from(edges)

nx.draw_networkx(G)
```

```python
deg_centra = nx.degree_centrality(G)
bet_centra = nx.betweenness_centrality(G)
close_centra = nx.closeness_centrality(G)
eigen_centra = nx.eigenvector_centrality(G)

df = pd.DataFrame([deg_centra, bet_centra, close_centra, eigen_centra])
ndf = df.transpose()
ndf.columns = ["Degree", "Betweeness", "Closeness", "Eigenvector"] 

print("Measures of Centrality")
ndf
```

```python
def unique_nodes(arr):
    n = set()
    for i in arr:
        n.add(i[0])
        n.add(i[1])
    
    return n
```

```python
start_node = 5
dfs_list = list(nx.dfs_edges(G, 5))
print(f"Edges traversed from starting with Node: {start_node} are {dfs_list}")
print(f"Nodes traversed by starting from Node {start_node} are {unique_nodes(dfs_list)}")
```

```python
start_node = 5
bfs_list = list(nx.bfs_edges(G, 5))
print(f"Edges traversed from starting with Node: {start_node} are {bfs_list}")
print(f"Nodes traversed by starting from Node {start_node} are {unique_nodes(bfs_list)}")
```
