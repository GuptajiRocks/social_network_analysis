```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty undirected graph
G = nx.Graph()

# Add nodes
G.add_node("A")
G.add_node("B")
G.add_node("C")

# Add edges
G.add_edge("A", "B")
G.add_edge("B", "C")
G.add_edge("C", "A")

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, edge_color='k', font_size=12)
plt.title("Simple NetworkX Graph")
plt.show()
```
